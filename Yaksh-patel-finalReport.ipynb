1)Precision:- What percent of your predictions were correct? The formula for precision is True Positives/True Positives+False Positives

2)Recall:- What percent of the positive cases did you catch? The formula for recall is  True Positives/True Positives +False Negatives.

3)F1 score:- What percent of positive predictions were correct? The formula for F1-score is 2*(Recall * Precision) / (Recall + Precision) .

4)Support:- Support is the number of actual occurrences of the class in the specified dataset. Imbalanced support in the training data may indicate structural weaknesses in the reported scores of the classifier and could indicate the need for stratified sampling or rebalancing. Support doesnâ€™t change between models but instead diagnoses the evaluation process.

To display the results I have used confusion matrix and the ROC curve. The confusion matrix is a table which is often used to describe the performance of the classification model on a set of test data for which the true values are known. It has four sides which is denoted as true positives(TP), true negatives(TN), false positives(FP) and false negatives(FN).

1)True positive (TP): correct positive prediction

2)False positive (FP): incorrect positive prediction

3)True negative (TN): correct negative prediction

4)False negative (FN): incorrect negative prediction

ROC curve is mainly used as a graph that summarizes the overall performance of a classifier over all possible thresholds. It is generated by plotting the True Positive Rate on y-axis Vs the False Positive Rate on x-axis.

## Discussions

I thought comparing this four classifiers to gather the obeservations and to find a reason why one's performaing better than other was important.

1) Decision tree Vs Random forest:- The Random Forest Algorithm concates the output of more than one randomly selected Decision Trees to generate the final output. Each node withinside the choice tree works on a random subset of functions to calculate the output. This process of combining the output of more than one individual models is called Ensemble Learning. As per the oberservations random forest is good for circumstances when we have a very large dataset, and interpretability is not a major issue. Random Forest has a higher training time than because It have multiple decision tree comapred to a single decision tree.

2) Gradient boosting Vs Stocastic gradient boosting:- As per the observations and theory In both gradient descent and stochastic gradient descent you have to update a set of parameters in different manner to minimize an error function. If the number of training samples are very large then using gradient descent will take more time because in every iteration when you are updating the values of the parameters and you are also running through the whole training set. Now In SGD, It will be faster because you use only one training sample and it starts improving right after from the first sample data. SGD mainly run much faster compared to gradient descent but the error function is not as well minimized as in the case of gradient descent. As I mentioned earlier gradient descent may have a disadvantage for model overfitting. For our dataset SGD is performing well compared to gradient boosting.

## data pre-processing

```python
def wordpre(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    
    return text
```

```python
#Applying the wordpre method to the dataset
Dataset1['Article']=Dataset1['Article'].apply(wordpre)
```

# Model Evaluation for Logistic regression

```python
#LogisticRegression
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression())])

Logisticmodel = pipe.fit(x_train, y_train)
prediction = Logisticmodel.predict(x_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction)*100,2)))
Logisticmodel_accuracy = round(accuracy_score(y_test, prediction)*100,2)
```
